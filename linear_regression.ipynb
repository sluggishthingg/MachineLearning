```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_excel('linear_reg.xlsx')
```

```python
data
```

```python
x=data['Height (cm)'].values.reshape(-1,1)
y=data['Weight (kg)'].values
```

```python
x
```

```python
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
y_pred
```

```python
mse=mean_squared_error(y,y_pred)
print(f'Mean squared error {mse}')
```

```python
plt.scatter(x,y,color='blue',label='Data_points')
plt.xlabel('Height(in cm)')
plt.ylabel('Weight (in kg)')
plt.title('Linear Regression: Height vs Weight')
plt.legend()
plt.show()
```

```python
plt.plot(y_pred,color='red',label='Regression Line')
plt.xlabel('Height(in cm)')
plt.ylabel('Weight (in kg)')
plt.title('Linear Regression: Height vs Weight')
plt.legend()
plt.show()
```

```python
print(f'Model Regression Coefficients:{model.coef_}')
print(f'Model Linear Regression :{model.intercept_}')
```
